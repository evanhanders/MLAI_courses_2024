We initially start with 5 values of N and 9 values of Y, so initially,

P(Y) = 5/14 and P(N) = 9/14.

```python
py = 5/14
pn = 9/14
H = - (py * np.log2(py) + pn * np.log2(pn))

# split on outlook:
# in sunny bin
py = 2/5
pn = 3/5
H_sunny = - (py * np.log2(py) + pn * np.log2(pn))
# in overcast bin
py = 4/4
pn = 0/4
H_overcast = - (py * np.log2(py) + pn * np.log2(pn))
H_overcast = 0 #because entropy is 0 when all examples are in the same class
# in rainy bin
py = 3/5
pn = 2/5
H_rainy = - (py * np.log2(py) + pn * np.log2(pn))

H_outlook = 5/14 * H_sunny + 4/14 * H_overcast + 5/14 * H_rainy

# split on temperature:
# in hot bin
py = 2/4
pn = 2/4
H_hot = - (py * np.log2(py) + pn * np.log2(pn))
# in mild bin
py = 4/6
pn = 2/6
H_mild = - (py * np.log2(py) + pn * np.log2(pn))
# in cool bin
py = 3/4
pn = 1/4
H_cool = - (py * np.log2(py) + pn * np.log2(pn))

H_temperature = 4/14 * H_hot + 6/14 * H_mild + 4/14 * H_cool

# split on humidity:
# in high bin
py = 3/7
pn = 4/7
H_high = - (py * np.log2(py) + pn * np.log2(pn))
# in normal bin
py = 6/7
pn = 1/7
H_normal = - (py * np.log2(py) + pn * np.log2(pn))

H_humidity = 7/14 * H_high + 7/14 * H_normal

# split on windy:
# in true bin
py = 3/6
pn = 3/6
H_true = - (py * np.log2(py) + pn * np.log2(pn))
# in false bin
py = 6/8
pn = 2/8
H_false = - (py * np.log2(py) + pn * np.log2(pn))

H_windy = 6/14 * H_true + 8/14 * H_false

#changes in entropy
print('Information gain for outlook:', H - H_outlook)
print('Information gain for temperature:', H - H_temperature)
print('Information gain for humidity:', H - H_humidity)
print('Information gain for windy:', H - H_windy)
```
